```python
#Stack Plot
import matplotlib.pyplot as plt
import pandas as pd

data = {'Year': [2015, 2016, 2017, 2018, 2019, 2020],
'Sales': [100, 120, 150, 170, 200, 220],
'Expenses': [70, 80, 90, 100, 110, 120]}
df = pd.DataFrame(data)

plt.stackplot(df['Year'], df['Sales'], df['Expenses'],
              labels=['Sales', 'Expenses'])
plt.xlabel('Year')
plt.ylabel('Amount ($)')
plt.title('Sales and Expenses Over Time')
plt.legend(loc='upper left')
plt.show()

#fill between
plt.fill_between(df['Year'], df['Sales'], df['Expenses'], #for feel between
alpha=0.5, label='Sales - Expenses')
plt.show()


fig, ax = plt.subplots()
ax.fill_between(df['Year'], df['Sales'], alpha=0.5, label='Sales')
ax.fill_between(df['Year'], df['Expenses'], alpha=0.5, label='Expenses')
plt.plot(df['Year'], df['Sales'], color="red", alpha=0.6)
plt.plot(df['Year'], df['Expenses'], color="green", alpha=0.6)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

months = np.arange(1, 13)
sales = np.array([10, 15, 20, 30, 20, 50, 30, 70, 80, 40, 100, 10])
expenses = np.array([5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60])

fig, ax = plt.subplots()

ax.plot(months, sales, color='green', label='Sales')
ax.plot(months, expenses, color='red', label='Expenses')


ax.fill_between(months, sales, expenses, where=sales > expenses,
                alpha=0.2, color='blue',interpolate=True)
ax.fill_between(months, sales, expenses, where=expenses > sales,
                alpha=0.2, color='yellow',interpolate=True)

ax.set_xlabel('Month')
ax.set_ylabel('Amount ($)')
ax.set_title('Sales and Expenses')
ax.legend()
plt.show()

plt.savefig('MySamplePlot.png', format='png') #for save
```

```python
'''============>RE(Regular Expressions) & Multi-Threding<================='''
```

```python
[] ->A set of characters                   ->"[a-m]"
\  ->Signals a special sequence            ->"\d"
.  ->Any Charte(Except newline char)       ->"he..o"
^  ->Strats With                           ->"^hello"
$  ->Ends With                             ->"planet$"
*  ->Zero or more occurrences              ->"he.*o"

?  ->Zero or one occurrences               ->"he.?o"
{} ->Exactly the specified number of occ   ->"he.{2}o"
|  ->Either or                             ->"falls|stays"
() ->one or more occurrences              
```

```python
#[] Square Brackets
import re

txt = "The rain in Spain"
x = re.findall("[a-m]", txt)
print(x)
x = re.findall("[^a-m]", txt)
print(x)
x = re.findall("[^a-m,' ']", txt)
print(x)
x = re.findall("[^1-4,' ']", txt)
print(x)
```

```python
#\ Backslash
import re

txt = "That will be 59 dollars"
#Find all digit characters:
x = re.findall("\d", txt)
print(x)

txt = "That will be .59 dollars."
x = re.findall("\.", txt)
print(x)

```

```python
#^ Caret (Start With)
import re
txt = "hello54 hello56 planet"
#Check if the string starts with 'hello':
x = re.findall("^hello", txt)
if x:
 print("Yes, the string starts with 'hello'")
 print(x)
else:
 print("No match")

txt = "hello4 hello65 planet"
#Check if the string starts with 'hello':
x = re.findall("^hello6", txt)
if x:
 print("Yes, the string starts with 'hello'")
 print(x)
else:
 print("No match")
```

```python
#$ Dollar (end with)
import re
txt = "hello planet"
#Check if the string ends with 'planet':
x = re.findall("planet$", txt)
if x:
 print("Yes, the string ends with 'planet'")
else:
 print("No match")

txt = "hello planet5"
#Check if the string ends with 'planet':
x = re.findall("t$", txt)
if x:
 print("Yes, the string ends with 'planet'")
else:
 print("No match")


txt = "hello planet"
#Check if the string ends with 'planet':
x = re.findall("t$", txt)
if x:
 print("Yes, the string ends with 'planet'")
else:
 print("No match")

```

```python
#? Question mark
import re
txt = "hello planet"
#Search for a sequence that starts with "he", followed by 0 or 1 (any)
#character, and an "o":
x = re.findall("he.?o", txt)
print(x)

txt = "helo planet"
x = re.findall("he.?o", txt)
print(x)
```

```python
#{m,n}- Braces
import re
txt = "hello planet"
#Search for a sequence that starts with "he", followed excactly 2 (any)
#characters, and an "o":
x = re.findall("he.{2}o", txt)
print(x)

txt = "hello planet"
x = re.findall("he.{1}o", txt)
print(x)

txt = "helllo planet"
x = re.findall("he.{3}o", txt)
print(x)


```

```python
#- OR
import re
txt = "The rain in Spain falls mainly in the plain!"
#Check if the string contains either "falls" or "stays":
x = re.findall("falls|stays", txt)
print(x)
if x:
 print("Yes, there is at least one match!")
else:
 print("No match")


txt = "The rain in Spain falls mainly in the plain stays!"
x = re.findall("falls|stays", txt)
print(x)
if x:
 print("Yes, there is at least one match!")
else:
 print("No match")
```

```python
#()- Group
import re
email = "rohit@yadav.com"
pattern = r"([a-z]+)@([a-z]+)\.com"
# Apply the pattern and extract the capture groups
match = re.match(pattern, email)
print(match.group())

if match:
 username = match.group(1)
 domain = match.group(2)
 print("Username:", username)
 print("Domain:", domain)
else:
 print("Email address is not valid")
```
